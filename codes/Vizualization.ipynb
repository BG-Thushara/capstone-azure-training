## Insights and Visualization

```
product.where(col('frq_of_product') == product.select(max(col('frq_of_product'))).collect()[0][0]).select('Product_ID','frq_of_product').show()
```

- Therfore, **PROD003** is product on demand, or most sold product.

```
sales.groupBy('customer_id').agg(sum('sale_amount').alias('sale_amount')).select(max(col('sale_amount'))).display()
```

```
sales.groupBy('customer_id').agg(sum('sale_amount').alias('sale_amount')).filter(col('sale_amount')==3599).display()
```

- The customer posessing **2584 - customer id** is the most loyal customer in terms of sales.

```
sales.groupBy('customer_id') \
     .agg(sum('sale_amount').alias('sale_amount')) \
     .orderBy(col('sale_amount').desc()) \
     .limit(10).display()
```

```
sales.groupBy('store_id') \
     .agg(sum('sale_amount').alias('sale_amount')) \
     .orderBy(col('sale_amount').desc()) \
     .limit(5).display()
```

```
sales.groupBy('sale_date').agg(sum('sale_amount').alias('sale_amount')).select(max(col('sale_amount'))).display()
```

```
sales.groupBy('sale_date').agg(sum('sale_amount').alias('sale_amount')).filter(col('sale_amount')==1493).display()
```

- on a whole, 2021-05-29 is the day, when the best sales happened.

```
inventory.groupBy('Product_ID').agg(sum('Inventory').alias('Inventory')).orderBy(desc(col('Inventory'))).display()
```

- **PROD014** is the most available product.
